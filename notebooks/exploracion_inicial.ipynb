El conjunto de datos cuenta con 3008 registros/filas y 28 columnas(2 str y 26 int).

```python
cantidad_nulos = df_INAH_visitas.isnull().sum()
print(f'Cantidad de valores nulos: {cantidad_nulos}')
```

```python
cantidad_duplicados = df_INAH_visitas.duplicated().sum()
print(f'Cantidad de registros/filas duplicadas: \n{cantidad_duplicados}')
```

No se identificaron valores nulos y duplicados.
